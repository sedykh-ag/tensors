```python
import numpy as np
from numpy import linalg as LA
```

```python
n = 10 # размерность ножек
d = 4  # кол-во ножек
f = lambda i1, i2, i3, i4: np.sin(i1 + i2 + i3 + i4)

A = np.fromfunction(f, (n, n, n, n), dtype=float)
A.shape
```

```python
def tt_decomp(X, n, d, delta=1e-4):
# подразумевается, что у X все ножки по n
    
    if len(A.shape) == d:
        return X
    
    else:
        X = X.reshape((-1, n))
        U, S, Vh = LA.svd(X)

        r = np.sum(S > delta)

        X_r = U[:, :r] @ np.diag(S[:r]) @ Vh[:r, :]
        tt_decomp(X_r)
```

```python
A_r = tt_decomp(A, n=n, d=d)
```
